```python
def to_structured(cleaned):
    structured = []
    
    current_key = None
    
    for item in cleaned:
        # detect key (biasanya Title Case / known field)
        if item.lower() in [
            'sensing method', 'model', 'sensing distance',
            'light source (wavelength)', 'power supply voltage',
            'current consumption', 'control output'
        ]:
            current_key = item
        else:
            if current_key:
                structured.append({
                    "section": current_key,
                    "content": item
                })
    
    return structured
```
